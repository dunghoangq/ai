# Neural Networks - Single Layer NN

## 1. Why Neural Networks Need Multiple Neurons

- XOR problem again.
- One neuron = one feature detector -> how to recognise an object with just one feature???
- Input -> one neuron -> one scaler.
- A layer ouput is called an **embedding**, **latent representation**, or **feature representation**

````python
import numpy as np

x = np.array([1.0, 2.0, 3.0])

weights = [
    np.array([1.0, 0.0, -1.0]),
    np.array([0.5, 1.0, 0.5]),
    np.array([-1.0, 2.0, 1.0])
]

biases = [0.0, 1.0, -2.0]

outputs = []

for w, b in zip(weights, biases):
    z = np.dot(w, x) + b
    outputs.append(z)

outputs = np.array(outputs)

print(outputs)
````

## 3. Understanding Learned Features

- A NN doesn't memorise patterns, it learns a hierarchy of features.
- One neuron learns how much the input resemble the pattern encoded by the neuron's weight.

````python
import numpy as np

X = np.array([
    [2, 1],
    [3, 1],
    [-2, -1],
    [-3, -1]
])

W = np.array([
    [1, -1],
    [1,  1]
])

features = X @ W

print(features)
````

## 3. Matrix Multiplication

- Each neuron has its weight vector, each is a question. Input vectors have to answer all these questions at once.
- Each neuron is a quality scanner. All scanners scan one product simultaneously.

```python
output = x @ W
```

- Batch is actually stacking all inputs x into one matrix X.

No numpy

```python
def matmul(A, B):
    m = len(A)
    n = len(A[0])
    if n != len(B):
        return
    p = len(B[0])
    result = [[0]*m]*p
    for i in range(m):
        for j in range(n):
            for k in range(p):
                result[i][k] += A[i][j] * B[k][j]
    return result
```

## 4. Linear Transformations

- Vector space.
- A Matrix moves every vector in the space simultaneously, reshape and rotate the space so properties/features are easier to process.
- Transformations:
  - Stretching
  - Shrinking
  - Rotation
  - Reflection
- Define a new coordinate system.
- Coordinates of the input expressed in terms of learned features.
- One transformation doesn't learn anything -> Many transformations = many layers.
  - e.g., layer 1 = detect edges, layer 2 = combine edges to shapes, layer 3 = combines shapes to objects.
- Not about numerical multiplication, it's about changing the geometry of information.

## 5. High-Dimensional Geometry

- Each object is a point in (high-dimensional) space.
- How close points are is how similar they are.
- Cosine Similarity is used in every model:
  - embeddings
  - RAG
  - retrieval
  - semantic search
  - recommendation systems
- Training -> Clusterings -> Neighbourhoods
- Model discovers: King - Man = Royalty, then King - Man + Women = Queen
- The **curse of dimensionality** 
  - roughly same distance -> Nearest neighbour search is harder.
  - nearly orthogonal -> meaningful similarity stands out clearly.
  - most volume lives near the surface -> can be exploited for optimisation.
- **Manifold**, realworld data doesn't cover the whole vector space, it is what learning discovers.
- **Latent space** = hidden layers in NN.
- Attention, meaning changes with context.
- Everything is **Nearest-Neighbour Search**.

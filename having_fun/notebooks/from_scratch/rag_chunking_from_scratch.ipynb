# RAG - Chunking From Scratch

Convert pieces inside a document into vectors.
Let the embedding model encode those vectors, because the model cannot search directly inside documents, because a document contains multiple ideas and subjects.
-> Chunking.

It's like mixing orange juice, coffee, soda,... in a bucket, the model will average all the flavours.

If user asks for a coffee, it returns a cup of coffee with orange flavour, added with carbonhydrate gas of soda.

**BUT**, if the chunk size if small, it'd lose the context.


> **Trade off**: Context vs. Precision
> Reasonable chunks contain `200-1000` tokens each, depending on applications.

A token ~ 0.75 word.

**So what is the problem when fixing chunk size?**

e.g., chunk size = 4

[A B C D]
[E F G H]
[I J K L]

What if we need [D E], neither the first two chunks contain both D and E.

> This is called **Boundary Loss**. One of the biggest problems in RAG.

To solve this problem, we use **Sliding chunk**.

All parameters now are:

- `chunk_size`
- `window_step`
- `overlap` := repeated part between two chunks.

> **Trade off**: Storage vs. Recall.
> Better start with `chunk_size = 500` tokens and `overlap = 50-100` tokens. Then tune experimentally.

## Init

```python
import numpy as np
```

## Pipeline

```python
tokens = [i for i in range(10)]

chunk_size = 4
overlap = 2
step = chunk_size - overlap

print(tokens)
```

```python
chunks = []

for start in range(0, len(tokens), step):
    end = start + chunk_size
    chunk = tokens[start:end]

    if not chunk:
        break
    chunks.append(chunk)

    if end >= len(tokens):
        break
```
